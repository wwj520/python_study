## 1.基础使用
#### 无限循环

```python
from tenacity import retry
# 没有任何参数无限循环
@retry
def never_gonna_give_you_up():
    print("Retry forever ignoring Exceptions, don't wait between retries")
    raise Exception
never_gonna_give_you_up()
```

#### 停止重试

```python
from tenacity import retry, stop_after_attempt
# stop_after_attempt(3)：重试3次
@retry(stop=stop_after_attempt(3))
def stop_after_3_attempts():
    print("Stopping after 3 attempts")
    raise Exception
stop_after_3_attempts()
```

#### 规定多久后停止

```python
from tenacity import retry, stop_after_attempt, stop_after_delay
# 0.5s后同停止重试
@retry(stop=stop_after_delay(0.5))
def stop_after_5_seconds():
    print("Stopping after 5 seconds")
    raise Exception
stop_after_5_seconds()
```

#### 多条件判断

```python
from tenacity import stop_after_attempt, stop_after_delay

# 2s后停止或重试5次
@retry(stop=(stop_after_delay(2) | stop_after_attempt(5)))
def stop_after_10_s_or_5_retries():
    print("Stopping after 10 seconds or 5 retries")
    raise Exception

stop_after_10_s_or_5_retries()
```

## 2. 重试的时间间隔 

#### 固定时间间隔重试

```python
from tenacity import wait_fixed
# 延迟 2 秒再进行下一次重试
@retry(wait=wait_fixed(2))
def wait_2_s():
    print("Wait 2 second between retries")
    raise Exception

wait_2_s()
```

#### 随机时间间隔

```python
from tenacity import  wait_random
# 2-5s 随机时间间隔
@retry(wait=wait_random(min=2, max=5))
def wait_random_1_to_2_s():
    print("Randomly wait 1 to 2 seconds between retries")
    raise Exception

wait_random_1_to_2_s()
```
